```python
veg = pd.read_csv("C:/Users/JulienPaulet/Desktop/Open_Classroom/P3_Realisez_une_etude_de_sante_publique/Donnees/Bilan_alimentaire/pour_sql/vege_q16.csv", encoding='utf-8')

ani = pd.read_csv("C:/Users/JulienPaulet/Desktop/Open_Classroom/P3_Realisez_une_etude_de_sante_publique/Donnees/Bilan_alimentaire/pour_sql/ani_q16.csv", encoding='utf-8')


# Ajout de la variable origin

ani["Origine"] = "animal"

veg["Origine"] = "vegetal"


# On regroupe veg et ani en un unique dataframe, via une union

temp_df = ani.append(veg)


# On supprime les anciens df ani et veg

del ani, veg

#On renomme les colonnes avec accents : 
#temp_df.rename(index=str, columns = {'Année' : 'Annee', 'Élément' : 'Element'})


# Transformation de temp_df en table pivot

data_q16 = temp_df.pivot_table(

    index=['Code Pays','Pays','Code Produit','Produit','Année','Origine'],

    columns = ["Élément"], values=["Valeur"])

#On renomme les colonnes comme précisé dans la consigne
data_q16.columns = ['dispo_alim_kcal_p_j', 'dispo_alim_tonnes', 'dispo_mat_gr', 'dispo_prot']

#On passe la colonne dispo_alim_tonnes en tonnes (unité de base en kg)
data_q16['dispo_alim_tonnes'] = data_q16['dispo_alim_tonnes']/1000

data_q16 = data_q16.reset_index()

data_q16.columns = ['code_pays', 'pays', 'code_produit', 'produit', 'annee', 'origine', 'dispo_alim_kcal-p-j', 'dispo_alim_tonnes', 'dispo_mat_gr', 'dispo_prot']

data_q16
```
