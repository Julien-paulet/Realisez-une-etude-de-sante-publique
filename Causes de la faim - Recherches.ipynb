```python
#-*- coding: utf-8 -*

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import randint, seed
```

```python
#Table population
pop = pd.read_csv("C:/Users/JulienPaulet/Desktop/Open_Classroom/P3_Realisez_une_etude_de_sante_publique/Donnees/Bilan_alimentaire/FAOSTAT_data_10-25-2018-1.csv", encoding='utf-8')

pop = pop[pop['Produit'] == 'Population']
pop = pop.pivot_table(index=['Pays'], columns='Produit', values='Valeur')
pop['Population'] = pop['Population']*1000



#On va ici calculer le ratio d'importations et d'exportations / Production : 

data = pd.read_csv("C:/Users/JulienPaulet/Desktop/Open_Classroom/P3_Realisez_une_etude_de_sante_publique/Donnees/causes_faim/imports_exports.csv", encoding='utf-8')

data = data.pivot_table(index = ['Pays', 'Produit'], columns = 'Élément', values = 'Valeur')
data.columns = ['Exports', 'Imports', 'Prod']
data = data.reset_index().groupby('Pays')

data = data.sum()

data_ratio = data

data_ratio['ratio exports/prod'] = data_ratio['Exports']/data_ratio['Prod']*100
data_ratio['ratio imports/prod'] = data_ratio['Imports']/data_ratio['Prod']*100

data_ratio = data_ratio.sort_values('ratio exports/prod', ascending = False)
data_ratio = data_ratio.round(1)[0:20]


#On merge les deux tables pour pouvoir calculer la disponibilité intérieure par habitant pour chaque pays : 

data_pop = pd.merge(data, pop, left_on = 'Pays', right_on = 'Pays')

#Et on calcule le ratio prod/population (En tonnes par habitant) : 
prod_pop = data_pop
prod_pop['ratio prod/pop'] = prod_pop['Prod']/prod_pop['Population']*1000
prod_pop = prod_pop.sort_values('ratio prod/pop', ascending=True)
prod_pop = prod_pop.round(1)[0:20]
```

```python
#Df sur les pertes et la nourriture pour chaque pays :
data_pn = pd.read_csv("C:/Users/JulienPaulet/Desktop/Open_Classroom/P3_Realisez_une_etude_de_sante_publique/Donnees/causes_faim/pertes_nourritures.csv", encoding='utf-8')

data_pn = data_pn.pivot_table(index=['Pays'], columns=['Élément'], values='Valeur')

#On met aux bonnes unités ; Population en population totale et nourritures, pertes, en tonnes
data_pn['Population totale'] = data_pn['Population totale']*1000
data_pn['Nourriture'] = (data_pn['Nourriture']*1000).round(0)
data_pn['Pertes'] = (data_pn['Pertes']*1000).round(0)

#On calcule ensuite la population totale : 
pop_tot = data_pn['Population totale'].drop(index='Chine, continentale').sum()

#On calcule ensuite la somme de pertes totale (en tonnes) : 
pertes_tot = data_pn['Pertes'].drop(index='Chine, continentale').sum()

#On calcule la somme de nourriture totale (en tonnes) : 
nourriture_tot = data_pn['Nourriture'].drop(index='Chine, continentale').sum()

#Puis enfin le ratio perte/nourriture
ratio_pn = (pertes_tot/nourriture_tot*100).round(0)
```

```python
print('La perte totale (en tonnes) est de : {} au niveau mondial'.format(pertes_tot))

print('------------------------')

print('La nourriture totale (en tonnes) est de : {} au niveau mondial'.format(nourriture_tot))

print('------------------------')

print('La ratio pertes/nourriture est de : {} %'.format(ratio_pn))
print('Ce dernier ratio ne prend pas en compte les pertes des ménages directement, le chiffre est donc bien plus élevé en réalité.')
```
