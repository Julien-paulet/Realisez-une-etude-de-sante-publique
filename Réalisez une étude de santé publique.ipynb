```python
"""Voici l'équation : 
    Production = Disponibilité intérieure + Exports - Imports - Variation de stocks = Semences + Nourritures + Aliments Animaux + Autres + Traitement + Pertes + Exports - Imports - Variation de stocks"""
```

```python
#test sur des valeurs réelles : 
#Production
data_france = data[data['Pays'] == 'France']
data_france_ble = data_france[data_france['Produit'] == 'Blé']

production_france = data_france_ble[data_france_ble['Code Élément'] == 5511]
somme_production_france_ble = np.sum(production_france['Valeur'] * 1000)
print("La production de blé en France est de : {} tonnes de blé".format(somme_production_france_ble))

#Dispo int
z_var_stocks = data_france_ble[data_france_ble['Code Élément'] ==  5072 ]
z_import = data_france_ble[data_france_ble['Code Élément'] ==  5611 ]
z_export = data_france_ble[data_france_ble['Code Élément'] ==  5911 ]
z_dispo_int = data_france_ble[data_france_ble['Code Élément'] ==  5301 ]

somme_dispoint_data_france_ble = (np.sum(z_dispo_int['Valeur']) + np.sum(z_export['Valeur']) - np.sum(z_import['Valeur']) - np.sum(z_var_stocks['Valeur'])) *1000

print("La disponibilité intérieure (moins les imports et la variation de stocks, et en ajoutant les exports) est de : {} tonnes".format(somme_dispoint_data_france_ble))

#Autres 
x_semences = data_france_ble[data_france_ble['Code Élément'] ==  5527 ]
x_nourritures = data_france_ble[data_france_ble['Code Élément'] ==  5142 ]
x_aliments_animaux = data_france_ble[data_france_ble['Code Élément'] ==  5521 ]
x_autres = data_france_ble[data_france_ble['Code Élément'] ==  5154 ]
x_traitement = data_france_ble[data_france_ble['Code Élément'] ==  5131 ]
x_pertes = data_france_ble[data_france_ble['Code Élément'] ==  5123 ]
x_var_stocks = data_france_ble[data_france_ble['Code Élément'] ==  5072 ]
x_import = data_france_ble[data_france_ble['Code Élément'] ==  5611 ]
x_export = data_france_ble[data_france_ble['Code Élément'] ==  5911 ]


somme_autre_data_france_ble = (np.sum(x_semences['Valeur']) + np.sum(x_nourritures['Valeur']) + np.sum(x_aliments_animaux['Valeur']) + np.sum(x_autres['Valeur']) + np.sum(x_traitement['Valeur'])  + np.sum(x_pertes['Valeur']) - np.sum(x_import['Valeur']) + np.sum(x_export['Valeur']) - np.sum(x_var_stocks['Valeur'])) *1000

print("La troisième partie de l'équation nous donne : {} tonnes de blé".format(somme_autre_data_france_ble))

if somme_production_france_ble == somme_dispoint_data_france_ble == somme_autre_data_france_ble :
    print("On obtient le même résultat pour chacun des calculs, l'équation est donc vérifiée.")
else : 
    print("L'équation n'est pas vérifiée.")
```

```python
data_usa = pd.read_csv("C:/Users/JulienPaulet/Desktop/Open_Classroom/P3_Realisez_une_etude_de_sante_publique/Donnees/Bilan_alimentaire/FAOSTAT_usa_q13.csv", encoding='utf-8')

data_usa = data_usa.drop(columns=['Élément','Code Élément','Code Pays','Pays', 'Code Année', 'Année', 'Symbole', 'Description du Symbole','Domaine', 'Code Domaine'])

data_usa = pd.DataFrame.pivot_table(data_usa, index='Code Produit', columns=['Unité'])

data_usa_sum = data_usa['Valeur'].sum().sum()

print("La quantité (en milliers de tonnes) de produits utilisés pour des aliments pour animaux est de : {}".format(data_usa_sum))
print("--------------------------------")

#Calcul du 10% de production en moins, soit 10% d'utilisation de nourriture en moins 

reduc_data_usa = data_usa_sum*0.90

print("Avec une réduction de 10% de production de produits de type animal, on obtient une utilisation pour l'alimentation animale de : {} milliers de tonnes".format(reduc_data_usa.round(0)))
print("--------------------------------")

#On doit maintenant calculer l'impact de cette réduction sur l'utilisation des céréales :

#On ajoute la colonne is_cereale de la question 11 : 

#Le df porte uniquement sur les céréales, afin de récupérer les codes produits
cereales = pd.read_csv("C:/Users/JulienPaulet/Desktop/Open_Classroom/P3_Realisez_une_etude_de_sante_publique/Donnees/Bilan_alimentaire/FAOSTAT_cereales.csv", encoding='utf-8')

cereales = cereales.drop(columns = ['Code Pays', 'Code Année', 'Année', 'Symbole', 'Description du Symbole','Domaine', 'Code Domaine'])
cereales = pd.DataFrame.pivot_table(cereales, index='Pays', columns=['Produit', 'Code Produit'])
#cereales = cereales.rename(index = str, columns={'Disponibilité alimentaire en quantité (kg/personne/an)': "Code produit"})
cereales = cereales['Valeur']
cereales.columns

#On ajoute une colonne en booléen qui indiquera si le produit est une céréale ou non

#Création d'une liste vide où l'on va mettre le résultat de la boucle
result = []

#Boucle for qui renvoie du booléen si le code produit correspond à un produit céréale
for index in data_usa.iterrows():
    if index[0] == 2511 or index[0] == 2513 or index[0] == 2514 or index[0] == 2515 or index[0] == 2516 or index[0] == 2517 or index[0] == 2518 or index[0] == 2520 or index[0] == 2805: 
        cereale = True 
    else:
        cereale = False
#On ajoute alors les valeurs directement dans la liste créée plus haut
    result.append(cereale)  

#On définit la liste en dataframe avec comme index les mêmes que pour data_nourriture_alim, et en  multi-colonnes comme pour data_nourriture_alim
result_df = pd.DataFrame(result, index=data_usa.index, columns=[['Valeur'],['is_cereal']])

#Enfin, on définit le df nourriture_alim en faisant une jointure des deux df
data_usa = pd.merge(data_usa, result_df, left_on='Code Produit', right_on='Code Produit')


#On applique un masque permettant d'obtenir la somme des produits céréales :

data_usa_cereal = data_usa[data_usa['Valeur','is_cereal'] == True]

data_usa_cereal_sum = data_usa_cereal['Valeur','Milliers de tonnes'].sum().sum()

print("La quantité de céréales utilisées pour l'alimentation animale est de : {} milliers de tonnes".format(data_usa_cereal_sum))
print("--------------------------------")

#Calcul du ratio cereale/total pour l'alimentation animale : 
ratio_usa_cereal = data_usa_cereal_sum/data_usa_sum

print("Les céréales constituent {} % des produits utilisés pour l'alimentation animales".format((ratio_usa_cereal*100).round(0)))
print("--------------------------------")

#On prend ensuite ce pourcentage et on l'applique sur la réduction pour trouver le nombre de milliers de produits céréales :
reduc_cereal_usa = reduc_data_usa*ratio_usa_cereal

#On soustrait alors le résultat à data_usa_cereal_sum pour avoir le nombre de tonnes économisées (on multipliera par 1000 car l'unité actuelle est le millier de tonnes)
reduc_cereal_usa = (data_usa_cereal_sum - reduc_cereal_usa)*1000

print("On libérerait {} tonnes de céréales pour une réduction de 10% de la production d'aliments de type animal".format(reduc_cereal_usa.round(0)))
```
